```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')   # word_tokenize ke liye bhi zaroori hota hai
```

```python
## stemming
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```python
## lemmatization
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(lemmatizer.lemmatize(word))
```

```python
## apply stopwords .lemmatize
import re
corpus=[]
for i in range(len(sentences)):
  review=re.sub( '[^a-zA-Z]',' ',sentences[i])
  review=review.lower()
  review=review.split()
  # Lemmatize words and remove stopwords
  processed_words=[lemmatizer.lemmatize(word)for word in review if not word in set(stopwords.words('english'))]
  # Join the processed words back into a single string (document)
  processed_sentence = ' '.join(processed_words)
  corpus.append(processed_sentence)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(binary=True)
```

```python
X=cv.fit_transform(corpus)
```
